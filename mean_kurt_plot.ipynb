```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nfs = [1, 2, 3, 4, 5, 6, 7, 8, 9]
yr = 3600 * 24 * 365.25
Tobs = 10 * yr 
freqs = np.array(nfs) / Tobs 
```

```python
WN = [7, 9, 11]

kurt_mean = []
kurt_std = []

for w in WN:
    data = np.genfromtxt(f'./plots_v2/kurt/gauss_gwb_{w}WN_curngwbA_kurt_100psrs.txt')

    kurt_mean.append(data[:, 0])
    kurt_std.append(data[:, 1])

kurt_mean = np.array(kurt_mean)
kurt_std = np.array(kurt_std)

plt.figure(figsize=(10, 6))
plt.title('Mean Kurtosis - curngwb')
for i, w in enumerate(WN):
    #plt.errorbar(freqs*1e9, kurt_mean[i], yerr=kurt_std[i], label=f'{w} WN', fmt='-o')
    plt.scatter(freqs*1e9, kurt_mean[i], label=f'{w} WN', s=50) 
plt.xlabel('Frequency [Hz]')
plt.ylabel('Mean_Kurtosis')
plt.grid(alpha=0.5)
plt.legend()
#plt.show()

plt.savefig('./plots_v2/kurt/kurt_curngwb.png')
```
